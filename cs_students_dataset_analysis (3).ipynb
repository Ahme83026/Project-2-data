```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cs_students_dataset.csv')
print(df.head())
```

```python
print("\nBASIC INFORMATION")
print("-" * 30)
print(f"Total students: {len(df)}")
print(f"Curriculum types: {df['Curriculum'].unique()}")
print("\nFirst 5 students:")
print(df.head())
```

```python
print("\nGPA STATISTICS BY CURRICULUM")
print("-" * 35)

grouped = df.groupby('Curriculum')['GPA']

print("\nAverage GPA:")
print(grouped.mean().round(2))

print("\nMedian GPA:")
print(grouped.median().round(2))

print("\nMinimum GPA:")
print(grouped.min().round(2))

print("\nMaximum GPA:")
print(grouped.max().round(2))

print("\nStandard Deviation:")
print(grouped.std().round(2))
```

```python
plt.figure(figsize=(8, 5))
means = grouped.mean().sort_values(ascending=False)
colors = ['#2ecc71', '#3498db', '#e74c3c', '#f39c12']
bars = plt.bar(means.index, means.values, color=colors[:len(means)])
plt.title('Average GPA by Curriculum', fontsize=14, fontweight='bold')
plt.xlabel('Curriculum')
plt.ylabel('Average GPA')
plt.ylim(0, 3.5)
for bar, value in zip(bars, means.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
             f'{value:.2f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Box Plot - GPA Distribution
plt.figure(figsize=(8, 5))

# Create box plot
df.boxplot(column='GPA', by='Curriculum')

# Customize
plt.title('GPA Distribution by Curriculum', fontsize=14, fontweight='bold')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Curriculum')
plt.ylabel('GPA')

plt.tight_layout()
plt.show()
```

```python
# Histogram - Overall GPA Distribution
plt.figure(figsize=(8, 5))

# Create histogram
plt.hist(df['GPA'], bins=10, color='teal', edgecolor='black', alpha=0.7)

# Add mean and median lines
mean_gpa = df['GPA'].mean()
median_gpa = df['GPA'].median()

plt.axvline(mean_gpa, color='red', linestyle='--', linewidth=2,
            label=f'Mean: {mean_gpa:.2f}')
plt.axvline(median_gpa, color='blue', linestyle='--', linewidth=2,
            label=f'Median: {median_gpa:.2f}')

# Labels and title
plt.title('Overall GPA Distribution', fontsize=14, fontweight='bold')
plt.xlabel('GPA')
plt.ylabel('Number of Students')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Create performance categories
def categorize_gpa(gpa):
    if gpa >= 3.0:
        return 'Excellent (3.0+)'
    elif gpa >= 2.5:
        return 'Good (2.5-2.99)'
    elif gpa >= 2.0:
        return 'Satisfactory (2.0-2.49)'
    else:
        return 'Needs Improvement (<2.0)'

df['Performance'] = df['GPA'].apply(categorize_gpa)

print("\n📊 PERFORMANCE BREAKDOWN")
print("-" * 30)

performance_counts = df['Performance'].value_counts()
print(performance_counts)

# Pie chart
plt.figure(figsize=(8, 6))
colors_perf = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']
plt.pie(performance_counts.values, labels=performance_counts.index,
        autopct='%1.1f%%', colors=colors_perf, startangle=90)
plt.title('Student Performance Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 50)
print("SUMMARY STATISTICS")
print("=" * 50)

print(f"\nOverall Statistics:")
print(f"  Mean GPA:     {df['GPA'].mean():.2f}")
print(f"  Median GPA:   {df['GPA'].median():.2f}")
print(f"  Std Dev:      {df['GPA'].std():.2f}")
print(f"  Min GPA:      {df['GPA'].min():.2f}")
print(f"  Max GPA:      {df['GPA'].max():.2f}")
curriculum_counts = df['Curriculum'].value_counts()
print(f"\nCurriculum Statistics:")
print(f"  Best:         {means.idxmax()} ({means.max():.2f})")
print(f"  Worst:        {means.idxmin()} ({means.min():.2f})")
print(f"  Most common:  {curriculum_counts.idxmax()} ({curriculum_counts.max()} students)")

print(f"\nPerformance:")
print(f"  Excellent:    {len(df[df['GPA'] >= 3.0])} students")
print(f"  Needs help:   {len(df[df['GPA'] < 2.0])} students")
```
